```python
import matplotlib.pyplot as plt

columns=["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]

plt.figure(figsize=(12, 10))

for index, column in enumerate(columns, 1):
    plt.subplot(3, 2, index)  
    plt.hist(df[column], bins=10, alpha=0.7, color='blue')
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()  
plt.show()

```

```python
import matplotlib.pyplot as plt
columns=["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]
values=df.loc[:,["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]]
plt.figure(figsize=(11,9))
plt.boxplot(values,labels=columns)
plt.show()
```
